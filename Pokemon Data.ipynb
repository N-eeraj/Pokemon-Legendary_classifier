### Legendary-Non-Legendary Graph

```python
legendary_ratio=[]

for i in types:
    legendary_ratio.append([i, False, len(globals()[i.lower()][globals()[i.lower()].Legendary==False])])
    legendary_ratio.append([i, True, len(globals()[i.lower()][globals()[i.lower()].Legendary==True])])

legendary_ratio_df=pd.DataFrame(legendary_ratio, columns=['Type', 'Legendary', 'Number'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sb

sb.catplot(data=legendary_ratio_df, kind="bar",  x="Type", y="Number", hue="Legendary")

plt.xticks(rotation=90)
plt.title('Legendary Pokemons by Type')
plt.xlabel('Types')
plt.ylabel('Number of Pokemons')
```

# Legendary Pokemon Classifier

```python
#spliting dataframe to features & outcome
x = df.drop(columns=['Type 1', 'Type 2', 'Generation', 'Legendary'])
y = df['Legendary']
```

```python
#splitting data to train & test data
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x.values, y.values)
```

### Training Model

```python
from sklearn.ensemble import RandomForestClassifier as rf

legendary_clf = rf(bootstrap=True)
legendary_clf.fit(x_train, y_train)
```

#### Accuracy of the Model
