### All-Rounder Pokemons

```python
hp75=df.HP.quantile(0.75)
atk75=df.Attack.quantile(0.75)
def75=df.Defense.quantile(0.75)
spd75=df.Speed.quantile(0.75)

allround_df=df[df.HP>hp75][df.Attack>atk75][df.Defense>def75][df.Speed>spd75]
allround_df
```

#### Types & count of Pokemon

```python
all_round_dict={}

for i in allround_df['Type 1']:
    if type(i)!=str:
        continue
    if i in all_round_dict:
        all_round_dict[i]+=1
    else:
        all_round_dict[i]=1
for i in allround_df['Type 2']:
    if type(i)!=str:
        continue
    if i in all_round_dict:
        all_round_dict[i]+=1
    else:
        all_round_dict[i]=1

all_round_dict
```

# Legendary Pokemon Classifier

```python
#spliting dataframe to features & outcome
x = df.drop(columns=['Type 1', 'Type 2', 'Generation', 'Legendary'])
y = df['Legendary']
```

```python
#splitting data to train & test data
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x.values, y.values)
```

### Training Model

```python
from sklearn.ensemble import RandomForestClassifier as rf

legendary_clf = rf(bootstrap=True)
legendary_clf.fit(x_train, y_train)
```

#### Accuracy of the Model

```python
'{}%'.format(legendary_clf.score(x_test, y_test)*100)
```
